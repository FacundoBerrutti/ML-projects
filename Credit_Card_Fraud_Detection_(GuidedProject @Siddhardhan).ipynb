````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

````

**LOAD DATA**

````python
credit_card_data = pd.read_csv('/content/creditcard.csv')
````

````python
# view head
credit_card_data.head()
````

````python
# view tail
credit_card_data.tail()
````

````python
credit_card_data.info()
````

````python
credit_card_data['Class'].value_counts() # 1-> Fraud ; 0->Normal
````

**SEPARATE DATA FOR ANALYSIS**

````python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
````

````python
print(legit.shape)
print(fraud.shape)
````

````python
credit_card_data.groupby('Class').mean()
````

**IMPROVE DATA DISTRIBUTION TO AVOID LEARNING MISTAKES**

````python
legit_sample = legit.sample(n=200)
````

````python
new_dataset = pd.concat([legit_sample,fraud], axis = 0) # axis = 0 because we are connecting files
````

````python
new_dataset['Class'].value_counts()
````

````python
new_dataset.groupby('Class').mean()
````

**DATA SPLITT**

````python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
````

````python
X.shape
````

````python
Y.shape
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y)
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

**MODEL TRAINING WITH LOGISTIC REGRESSION**

````python
model = LogisticRegression()
````

````python
model.fit(X_train, Y_train)
````

MODEL EVALUATION WITH ACCURACY SCORE

````python
#TRAINING DATA
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data is:', training_data_accuracy)
````

````python
#TEST DATA
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data is:', test_data_accuracy)
````
